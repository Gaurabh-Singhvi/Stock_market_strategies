```python
import nbimporter
import pandas as pd
import numpy as np
import time
import stock_list
from importlib import reload
reload(stock_list)
```

```python
final_list = stock_list.get_list_of_stocks()
```

```python
def screener_data_puller(stock):
    
    url = 'https://www.screener.in/company/' + str(stock) + '/consolidated'
    all_data = pd.read_html(url)
    
    shareholding_pattern = all_data[-2]
    shareholding_pattern['Stock Name'] = stock
    
    yearly_sales = all_data[1]
    yearly_sales['Stock Name'] = stock
    
    qtly_sales = all_data[0]
    qtly_sales['Stock Name'] = stock
    
    return shareholding_pattern, yearly_sales, qtly_sales
```

```python
a = []
b = []
c = []
error_list = []
j=0
for stock in final_list['SYMBOL'].unique().tolist():
    j = j+1
    try:
        if(j%15==0):
            time.sleep(30)
            print(stock)
            shareholding, yearly_result, qtly_result = screener_data_puller(stock)
            a.append(shareholding)
            b.append(yearly_result)
            c.append(qtly_result)

        else:
            print(stock)
            shareholding, yearly_result, qtly_result = screener_data_puller(stock)
            a.append(shareholding)
            b.append(yearly_result)
            c.append(qtly_result)
            
    except:
        error_list.append(stock) 
        continue

final_shareholding = pd.concat(a)
final_yearly_result = pd.concat(b)
final_qtly_result = pd.concat(c)
```
